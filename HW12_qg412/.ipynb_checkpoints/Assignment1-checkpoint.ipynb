**Several stations show a prominent annual periodicity. Identify the 4 **stations** (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)**

and

np.fft.rfftfreq() to create a meaningful x axis 

### Additional Notes from consultation with Prof Bianco:

What is Fourier Transformation? 

The theory behind it is that: So long as we have enough (infinite) sine graphs, we can construct any form of time series. So in fourier transformation, what it is doing is that it is trying to find how many sine graphs is needed to recreate the same time series as the one we currently observe. 

The whole purpose of this exercise is to identify the value of x (frequency) for which Y is highest, and subsequently find the highest 4 y-value for that identified value of x.

**Compressing 3d array into desired 2d array with only the stations and time.**

```python
mta_2d = np.nansum(mta_float, axis=1)
mta_2d.shape
```

**Running Fourier**

```python
power = (np.abs(np.fft.rfft(mta_2d , axis = 1))**2)
power.shape
```

```python
N = len(mta_2d[0])
print(N)
```

```python
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)

for i in range(len(mta_2d)):
    ax.plot(np.fft.rfftfreq(N, 1.0), power[i], 'o', ms=20)
    ax.plot(np.fft.rfftfreq(N, 1.0), power[i], '-')

pl.ylabel("Power", fontsize=20)
pl.xlabel("Frequency (1/week)", fontsize=20);
```

Figure 3: Plot showing power and frequency. Most of the data points are "squished" in the 0.0 region, with one point having extremely high power value. Trends are not visible probably due to the high outlier.

**Removing infinity to visualize plots better**

```python
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)

for i in range(len(mta_2d)):
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], power[i][1:], 'o', ms=10)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], power[i][1:], '-')

# changing the labels on the x ticks to see the period instead of frequency
ax.set_xticklabels([ "%s"%(1/power) for power in ax.get_xticks()], fontsize=20)
pl.xlabel("period (week)", fontsize=20)
pl.ylabel("Power", fontsize=20)
pl.title("Power spectrum and Period", fontsize=20);
```

Figure 4: Plot showing power and time period in weeks. After removing the outlier point at infinity, the trends are much more visible. There seems to be various peaks of power, with the highest before time period of 10 weeks followed by time peirod between 5 and 3 weeks. 

**Zooming in**

```python
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)

for i in range(len(mta_2d)):
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], power[i][1:], 'o', ms=10)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], power[i][1:], '-')

pl.xlim(0, 0.05)
# changing the labels on the x ticks to see the period instead of frequency
ax.set_xticklabels([ "%s"%(1/power) for power in ax.get_xticks()], fontsize=20)
pl.xlabel("period (week)", fontsize=20)
pl.ylabel("Power", fontsize=20)
pl.title("Power spectrum and Period", fontsize=20);
```

Figure 5: Plot showing zoomed in power specturm and period. The highest powered graphs seems to be concentrated around the 50th week period. 

**Defining frequency, the original x-axis**

```python
freq = np.fft.rfftfreq(N, 1.0)
freq.shape
```

Because fourier is a sine graph theory where there are oscillations that goes through y=0, we can literally count how many oscillation is it for time period 50, which in this case is 4.

Therefore it is meaningful to find out when frequency =4, what is the corresponding time period, which in this case, is 1/ freq4...

```python
1/freq[4]
```

This means that the trend repeats itself every 48.5 weeks (i think)

```python
#code credit: Mei @yg833
find = np.empty([600])
for i in range(len(power)):
    find[i] = power[i][4]
```
